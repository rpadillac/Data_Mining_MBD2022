```python
path = 'datasets/Canada.xlsx'
df = pd.read_excel(path,
                   sheet_name= "Canada by Citizenship",
                   skiprows= range(20),
                    skipfooter =2,
                    engine='openpyxl'
                      )
df
```

```python
df.head()
```

```python
print(df.columns)
```

```python
df.drop(['Type', 'Coverage', 'AREA', 'REG','DEV','DevName'],
        axis=1,
        inplace= True)
df.head
```

```python
#rename columns
df.rename(columns = {'OdName':'Country',
                     'AreaName': 'Continent',
                     'RegName': 'Region',                  
                    },
          inplace= True)

df.head()
          
```

## make a competitive study on top 5 countries

```python
top = df.sort_values(by = 'Totals',
                      ascending= False,
                      axis = 0)
top6= top.head(6)
top6
```

```python
top6 = top6[years].transpose()
top6
```

```python
line_graph_top6 = top6.plot(kind ='line', figsize=(15,10))
plt.title('Immigration to Canada', fontsize=15)
plt.ylabel('# of immigrants', fontsize=15)
plt.ylabel('Year', fontsize=15)
plt.text(7, 1180,'Reason', fontsize=18)
plt.show()
```

## Area plots (Staked line plot)

```python
top6.plot(kind= 'area',
          figsize=(15,20))

plt.title('StackedLinePlot', fontsize=18)
plt.show
```

```python
top6.plot(kind= 'area',
          figsize=(15,20),
         stacked = False) #to have transparency behind

plt.title('StackedLinePlot', fontsize=18)
plt.show
```

```python
top6.plot(kind= 'area',
          figsize=(15,20),
         stacked = False,
         alpha = 0.15) #color density

plt.title('StackedLinePlot', fontsize=18)
plt.show
```
