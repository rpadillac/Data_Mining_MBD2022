```python
#Download the 'anthropometry-dataset.csv' file (if you haven't already downloaded it).
#Open the .csv file using python.

path = 'datasets/antropometria-dataset.csv'
df= pd.read_csv(path)
df.head()
```

```python
x=df.columns.tolist()
x.sort()
x
```

```python
#Generate a pivot table showing the average weight of people by state.

st_wt = pd.pivot_table(df,
                      index = ['desc_ent'],
                      values = ['peso'],
                      aggfunc = 'mean',
                     )
st_wt
```

```python
#Calculate average weight of sex 1 and sex 2 by state.
st_wt = pd.pivot_table(df,
               index = 'desc_ent',
               values = ['peso'],
               aggfunc = ['mean','median','min'],
               columns = 'sexo',
              )
st_wt
```
